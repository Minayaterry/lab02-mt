```python
import pandas as pd
from pycaret.classification import *

# Cargar datos
df = pd.read_csv('diabetes.csv')

# Revisar las primeras filas
df.head()
```

```python
# Mostrar las primeras filas y la información general
df.head(), df.info()
```

```python
# Configurar PyCaret con la variable objetivo 'Outcome'
clf_setup = setup(df, target='Outcome', session_id=123)
```

```python
# Compararando modelos de clasificación y seleccionar el mejor
best_model = compare_models()
```

```python
# Evaluando el mejor modelo
evaluate_model(best_model)
```

```python
# Ajuste de hiperparámetros para mejorar el rendimiento del mejor modelo
tuned_model = tune_model(best_model)
```

```python
# Mostrar métricas después del ajuste
evaluate_model(tuned_model)
```

```python
# Predicciones con el modelo ajustado
predictions = predict_model(tuned_model)
print(predictions.head())
```
